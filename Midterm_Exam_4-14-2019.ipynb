### Q6: 以序號(連續數字)替換buyer_mobile的內容

```python
DataDF = pd.DataFrame(TransactionDataDF)


DataDF["rank"] = DataDF["buyer_mobile"].rank()

DataDF.sort_values("rank", inplace = True) 

i = 0
j = 0
objectM = DataDF["rank"][j]

while j < len(DataDF):
    if(DataDF["buyer_mobile"][j] == objectM):
        DataDF["buyer_mobile"][j] = str(i)
        j += 1
        continue
    DataDF["buyer_mobile"][j] = str(i+1)
    j += 1
    i += 1
    
    
count = len(DataDF.buyer_mobile)

ser = pd.Series(np.arange(count))

DataDF["buyer_mobile"] = ser

DataDF
```

### Q7: 將crdate字符串類型轉換為Datetime類型

```python
DataDF['crdate'] = pd.to_datetime(DataDF['crdate']) #.astype(str)

DataDF['crdate']
```

### Q8: 計算每位消費者最後一次消費的日期距離最後交易日(2016-12-31)的天數(Recency), 例如某消費者最後一次消費日是2016-12-28 14:40:43.350772則答案為2日), 經計算後產生一個名為RecencyDF的DataFrame, 其欄位為'buyer_mobile', 'Latest_Purchase_Date'以及'Recency'

```python
grouptimeDF=DataDF.groupby(['buyer_mobile'])['crdate'].max().reset_index()
Recency=TransactionEndingDate.day-grouptimeDF.crdate.map(lambda x:x.day)

Recencydata={'buyer_mobile':grouptimeDF.buyer_mobile,
             'Latest_Purchase_Date':grouptimeDF.crdate,
             'Recency':Recency}
RecencyDF=pd.DataFrame(Recencydata)
RecencyDF
```

### Q9: 計算每位消費者共消費次數(PurchaseTimes)

```python
OrderTimesDF=DataDF.groupby(['order_id', 'buyer_mobile']).size().reset_index(name='OrderTimes')
PurchaseTimesDF=OrderTimesDF.groupby(['buyer_mobile']).size().reset_index(name='PurchaseTimes')
PurchaseTimesDF
```

### Q10: 計算每位消費者消費次數統計值, 統計值含平均值(MeanTurnover), 標準差(StdTurnover),最小值(MinTurnover),最大值(MaxTurnover), 產生一一個名為FrequencyStatisticsDF的DataFrame, 其欄位為['buyer_mobile', 'MeanTurnover', 'StdTurnover', 'MinTurnover', 'MaxTurnover', 'PurchaseTimes']

```python
times=OrderTimesDF.groupby(['buyer_mobile'])['OrderTimes']
FrequencyStatisticsdata={'buyer_mobile':PurchaseTimesDF.buyer_mobile, 
                         'MeanTurnover':times.mean().reset_index().OrderTimes, 
                         'StdTurnover':times.std().reset_index().OrderTimes, 
                         'MinTurnover':times.min().reset_index().OrderTimes, 
                         'MaxTurnover':times.max().reset_index().OrderTimes, 
                         'PurchaseTimes':PurchaseTimesDF.PurchaseTimes}
FrequencyStatisticsDF=pd.DataFrame(FrequencyStatisticsdata)
FrequencyStatisticsDF
```

### Q11: 計算每位消費者共消費金額(TotalSpending)

```python
TotalSpendingDF=DataDF.groupby(['buyer_mobile'])['total'].sum().reset_index(name='TotalSpending')
TotalSpendingDF
```

### Q12: 計算每位消費者消費金額統計值, 統計值含平均值(MeanSpending), 標準差(StdSpending),最小值(MinSpending),最大值(MaxSpending), 產生一一個名為MonetaryStatisticsDF的DataFrame, 其欄位為['buyer_mobile', 'MinSpending', 'MaxSpending', 'MinTurnover', 'MaxTurnover', 'TotalSpending']

```python
grouptotalDF=DataDF.groupby(['order_id','buyer_mobile'])['total'].sum().reset_index(name='grouptotal')
times2=grouptotalDF.groupby(['buyer_mobile'])['grouptotal']

MonetaryStatisticsdata={'buyer_mobile':TotalSpendingDF.buyer_mobile, 
                        'MeanSpending':times2.mean().reset_index().grouptotal, 
                        'StdSpending':times2.std().reset_index().grouptotal,
                        'MinSpending':times2.min().reset_index().grouptotal,
                        'MaxSpending':times2.max().reset_index().grouptotal,
                        'MinTurnover':times.min().reset_index().OrderTimes, 
                        'MaxTurnover':times.max().reset_index().OrderTimes, 
                        'TotalSpending':TotalSpendingDF.TotalSpending}
MonetaryStatisticsDF=pd.DataFrame(MonetaryStatisticsdata)
MonetaryStatisticsDF
```

### Q13: 使用merge函數將RecencyDF, FrequencyStatisticsDF, MonetaryStatisticsDF整合成為一個DataFrame並命名為RFMsDF

```python
objectM=RecencyDF.merge(FrequencyStatisticsDF, left_on=None,right_on=None)
RFMsDF =objectM.merge(MonetaryStatisticsDF, left_on=None,right_on=None)

RFMsDF
```
